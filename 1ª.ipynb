```
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de las señales
amplitud = 1
frecuencia = 1
fase = 0

# Tiempos para la primera señal (más puntos)
tiempo1 = np.linspace(0, 2 * np.pi, 1000)  # 1000 puntos
senal1 = amplitud * np.sin(2 * np.pi * frecuencia * tiempo1 + fase)

# Tiempos para la segunda señal (menos puntos)
tiempo2 = np.linspace(0, 2 * np.pi, 200)  # 200 puntos
senal2 = amplitud * np.sin(2 * np.pi * frecuencia * tiempo2 + fase)


# Gráfica de las señales
plt.figure(figsize=(10, 6))
plt.plot(tiempo1, senal1, label='Señal 1 (100 puntos)', marker='o', markersize=3)
plt.plot(tiempo2, senal2, label='Señal 2 (20 puntos)', marker='s', markersize=8)


plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Dos señales sinusoidales con diferentes tamaños de paso')
plt.legend()
plt.grid(True)
plt.show()
```

```
# Calcula la Transformada de Fourier de ambas señales
transformada1 = np.fft.fft(senal1)
transformada2 = np.fft.fft(senal2)

# Frecuencias correspondientes a la transformada
frecuencias1 = np.fft.fftfreq(len(senal1), d=(tiempo1[1] - tiempo1[0]))
frecuencias2 = np.fft.fftfreq(len(senal2), d=(tiempo2[1] - tiempo2[0]))


# Gráfica de los espectros de frecuencia
plt.figure(figsize=(10, 6))
plt.plot(frecuencias1, np.abs(transformada1), label='Espectro de Señal 1')
plt.plot(frecuencias2, np.abs(transformada2), label='Espectro de Señal 2')


plt.xlabel('Frecuencia')
plt.ylabel('Amplitud')
plt.title('Espectros de Frecuencia de las Señales')
plt.xlim(-20,20)
plt.legend()
plt.grid(True)
plt.show()
```
